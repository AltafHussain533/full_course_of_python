```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df=sns.load_dataset("titanic")
```

```python
# Swarm
sns.swarmplot(data=df,x='age',y='fare')
```

```python
# Swarm
sns.swarmplot(data=df,x='survived',y='fare')
```

```python
# Swarm
sns.swarmplot(data=df,x='survived',y='age')
```

```python
# Swarm
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.catplot(x='pclass',y='fare',hue='embarked',kind='point',data=df)
```

```python
# relation of two numeric variables
sns.lmplot(data=df,x='fare',y='age')
```

```python
# relation of two numeric variables
sns.lmplot(data=df,x='age',y='fare')
```

```python
# relation of two numeric variables
sns.lmplot(data=df,x='age',y='fare',hue='pclass')
plt.ylim(0,300)
```

```python
# how to add a column and make a age group
sns.histplot(data=df,x='age')
```

```python
# grouping
df.columns
```

```python
df['age_groups']=pd.cut(df['age'],[0,10,15,30,50,80],labels=['kids','child','youth','middle aged','old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_groups")["survived"].mean().plot(kind="bar")
```

```python
df.groupby("age_groups")["fare"].mean().plot(kind="bar")
```

```python
sns.pairplot(df,hue='survived',vars=['age','fare','pclass'])
```

```python
sns.lmplot(data=df[df['fare']<300],x='age',y='fare')
```

```python
sns.lmplot(data=df[df['fare']<300],x='age',y='fare',hue='pclass')
```
