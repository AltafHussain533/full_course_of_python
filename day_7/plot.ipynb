# plotting in python

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
kashti=sns.load_dataset('titanic')
kashti
```

```python
kashti.info
```

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived',data=kashti)
```

```python
sns.countplot(x='survived',hue='sex',data=kashti)
```

```python
# scatter plot
plt.scatter(data=kashti,x="age",y="fare")
```

```python
sns.scatterplot(data=kashti,x="age",y="fare",hue="pclass")
```

```python
sns.barplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
sns.pairplot(kashti)
```

```python
g=sns.FacetGriid(kashti,hue='survived',col='pclass',margin_titles=True,
                palette=(1:"seagreen", 0:"gray"))
g=g.map(plt.scatter,'fare','age',edgecolor='W').add_legend():
```

```python
sns.stripplot(data=kashti,x='sex',y='fare',hue='class')
```

```python
# heat map
corr=kashti.corr()
plt.figure(figsize=(10,10))
```

```python
# heat map
corr=kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat map of kashti data")
```
