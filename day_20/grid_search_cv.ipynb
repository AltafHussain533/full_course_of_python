```python
# linear Regression
import pandas as pd
import seaborn as sns
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

# Load the dataset
df = sns.load_dataset('titanic')

# Handle missing values in 'age' and 'fare'
# Impute missing 'age' values with the mean of the column
df['age'] = df['age'].fillna(df['age'].mean())

# Select features and target
x = df[['age','pclass','sex','sibsp','parch','fare']]
y = df['survived']

# Convert categorical 'sex' column to numerical using one-hot encoding
# This replaces the 'sex' column with 'sex_male' and 'sex_female' columns with 0s and 1s
x = pd.get_dummies(x, columns=['sex'], drop_first=True) # drop_first avoids multicollinearity

# Define parameter grid
param_grid={'fit_intercept':[True,False]}

# Create a model
model = LinearRegression()

# GridSearchCV is generally not applicable for standard LinearRegression without additional parameters to tune (like regularization in Lasso/Ridge)
# object grid search cv (creating the model)
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='r2')

# Training the model
grid_search.fit(x, y)

# Output best parameters found
print("Best parameters found: ", grid_search.best_params_)
print("Best R^2 score: ", grid_search.best_score_)
```
